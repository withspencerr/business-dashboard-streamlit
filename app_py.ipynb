```python
# Interactive Business Dashboard Demo
# ç›®æ ‡ï¼šä½¿ç”¨ Streamlit + Plotly åˆ›å»ºä¸€ä¸ªç®€æ´çš„å•†ä¸šåˆ†æä»ªè¡¨ç›˜

!pip -q install streamlit plotly pandas

import pandas as pd
import numpy as np
import plotly.express as px
import streamlit as st

# ===== æ¨¡æ‹Ÿä¸šåŠ¡æ•°æ® =====
np.random.seed(42)
dates = pd.date_range(start="2023-01-01", periods=180)
regions = ["North", "South", "East", "West"]
products = ["Product A", "Product B", "Product C"]

data = []
for date in dates:
    for region in regions:
        for product in products:
            sales = np.random.randint(1000, 5000)
            profit = np.random.uniform(0.1, 0.3) * sales
            data.append([date, region, product, sales, profit])

df = pd.DataFrame(data, columns=["Date", "Region", "Product", "Sales", "Profit"])
df["Profit Margin"] = df["Profit"] / df["Sales"]

# ===== Streamlit ä»ªè¡¨ç›˜ä¸»é€»è¾‘ =====
st.title("ğŸ“Š Business Analytics Dashboard")
st.markdown("Explore revenue, profit, and product performance across regions.")

# ä¾§è¾¹æ è¿‡æ»¤
region_filter = st.sidebar.multiselect("Select Region(s):", options=df["Region"].unique(), default=df["Region"].unique())
product_filter = st.sidebar.multiselect("Select Product(s):", options=df["Product"].unique(), default=df["Product"].unique())

filtered_df = df[(df["Region"].isin(region_filter)) & (df["Product"].isin(product_filter))]

# ===== ä¸»å›¾è¡¨ =====
st.subheader("ğŸ’° Sales Trend Over Time")
sales_trend = filtered_df.groupby(["Date"], as_index=False)["Sales"].sum()
fig1 = px.line(sales_trend, x="Date", y="Sales", title="Total Sales Over Time", template="plotly_white")
st.plotly_chart(fig1, use_container_width=True)

st.subheader("ğŸ“ Regional Sales Distribution")
region_sales = filtered_df.groupby("Region", as_index=False)["Sales"].sum()
fig2 = px.bar(region_sales, x="Region", y="Sales", color="Region", title="Sales by Region", template="plotly_white")
st.plotly_chart(fig2, use_container_width=True)

st.subheader("ğŸ“¦ Product Performance")
product_perf = filtered_df.groupby("Product", as_index=False).agg({"Sales":"sum","Profit":"sum","Profit Margin":"mean"})
fig3 = px.bar(product_perf, x="Product", y="Sales", color="Profit Margin", title="Sales vs Profit Margin", template="plotly_white")
st.plotly_chart(fig3, use_container_width=True)

st.caption("Demo dashboard by Spencer Cai â€” MSBA @ WashU")
```
